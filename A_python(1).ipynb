```python
# ATM
import sys

# Mock Database: { 'account_num': {'pin': 'xxxx', 'balance': 0.0} }
users_db = {
    "101": {"pin": "1111", "balance": 500.0},
    "102": {"pin": "2222", "balance": 1200.0}
}

def create_account():
    print("\n--- Create New Account ---")
    acc_num = input("Enter a unique Account Number: ")

    if acc_num in users_db:
        print("Error: Account number already exists!")
        return

    new_pin = input("Set your 4-digit PIN: ")
    initial_deposit = float(input("Enter initial deposit amount: "))

    users_db[acc_num] = {"pin": new_pin, "balance": initial_deposit}
    print(f"Account {acc_num} created successfully!")

def login():
    print("\n--- Login to Giet Bank ---")
    acc_num = input("Account Number: ")

    if acc_num not in users_db:
        print("Account not found.")
        return None

    attempts = 3
    while attempts > 0:
        pin = input(f"Enter PIN ({attempts} attempts left): ")
        if users_db[acc_num]["pin"] == pin:
            print("Login Successful!")
            return acc_num
        else:
            attempts -= 1
            print("Incorrect PIN.")

    print("Security Alert: Too many failed attempts.")
    return None

def atm_interface(acc_num):
    user_data = users_db[acc_num]
    while True:
        print(f"\n--- Account: {acc_num} ---")
        print("1. Balance Inquiry\n2. Deposit\n3. Withdraw\n4. Logout")
        choice = input("Select Option: ")

        if choice == '1':
            print(f"Current Balance: ${user_data['balance']:,.2f}")
        elif choice == '2':
            amt = float(input("Deposit amount: "))
            user_data['balance'] += amt
            print("Deposit successful.")
        elif choice == '3':
            amt = float(input("Withdrawal amount: "))
            if amt <= user_data['balance']:
                user_data['balance'] -= amt
                print("Please take your cash.")
            else:
                print("Insufficient funds.")
        elif choice == '4':
            break

# Main App Loop
while True:
    print("\n=== WELCOME TO THE ATM SYSTEM ===")
    print("1. Login\n2. Create Account\n3. Shutdown System")
    main_choice = input("Select: ")

    if main_choice == '1':
        active_user = login()
        if active_user:
            atm_interface(active_user)
    elif main_choice == '2':
        create_account()
    elif main_choice == '3':
        print("System shutting down...")
        break
```

```python
#Student Gradebook

def calculate_average(scores):

    if not scores:
        return 0
    return sum(scores) / len(scores)

def get_letter_grade(average):
    if average >= 90: return "A"
    elif average >= 80: return "B"
    elif average >= 70: return "C"
    elif average >= 60: return "D"
    else: return "F"


def display_gradebook(gradebook):
    if not gradebook:
        print("\n[!] The gradebook is currently empty.")
        return

    print(f"\n{'STUDENT':<15} | {'AVERAGE':<10} | {'GRADE':<5}")
    print("-" * 35)
    for student, scores in gradebook.items():
        avg = calculate_average(scores)
        letter = get_letter_grade(avg)
        print(f"{student:<15} | {avg:<10.2f} | {letter:<5}")
    print("-" * 35)

def main():
    gradebook = {
        "Alice": [85, 92, 88],
        "Bob": [70, 65, 72]
    }

    while True:
        print("\n--- Student Gradebook System ---")
        print("1. View Gradebook")
        print("2. Add Student & Scores")
        print("3. Exit")

        choice = input("Select an option (1-3): ")

        if choice == '1':
            display_gradebook(gradebook)

        elif choice == '2':
            name = input("Enter student name: ").capitalize()
            try:
                score_input = input("Enter scores separated by commas: ")
                scores = [float(s.strip()) for s in score_input.split(',')]
                gradebook[name] = scores
                print(f"Added {name} successfully.")
            except ValueError:
                print("[!] Invalid input. Please enter numbers separated by commas.")

        elif choice == '3':
            print("Exiting gradebook. Goodbye!")
            break
        else:
            print("[!] Invalid choice, please try again.")

if __name__ == "__main__":
    main()
```

```python
#Password strength checker

import re

def password_strength(password):
    score = 0
    feedback = []

    if len(password) >= 8:
        score += 1
    else:
        feedback.append("Password is too short (minimum 8 characters).")


    if re.search(r"[a-z]", password) and re.search(r"[A-Z]", password):
        score += 1
    else:
        feedback.append("Include both uppercase and lowercase letters.")


    if re.search(r"\d", password):
        score += 1
    else:
        feedback.append("Include at least one number (0-9).")


    if re.search(r"[!@#$%^&*(),.?\":{}|<>+=-]", password):
        score += 1
    else:
        feedback.append("Include at least one special character (e.g., @, #, $).")

    return score, feedback

def main():
    print("--- Password Strength Checker ---")
    user_password = input("Enter a password to test: ")

    strength_score, suggestions = password_strength(user_password)


    strength_levels = {
        0: "Very Weak",
        1: "Weak",
        2: "Fair",
        3: "Good",
        4: "Strong"
    }

    print(f"\nStrength: {strength_levels[strength_score]}")

    if suggestions:
        print("Improvements needed:")
        for note in suggestions:
            print(f"- {note}")
    else:
        print("Great job! Your password meets all security criteria.")

if __name__ == "__main__":
    main()
```
